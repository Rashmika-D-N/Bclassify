```python
# 3. Setup ngrok for public URL so external models can reach this API
import pyngrok.ngrok as ngrok

print("=" * 60)
print("STEP 1: Setting up ngrok")
print("=" * 60)

# Replace with your actual ngrok auth token from https://dashboard.ngrok.com/get-started/your-authtoken
NGROK_AUTH_TOKEN = "YOUR_NGROK_AUTH_TOKEN_HERE"

if NGROK_AUTH_TOKEN == "YOUR_NGROK_AUTH_TOKEN_HERE":
    print("⚠ WARNING: Replace NGROK_AUTH_TOKEN with your actual token!")
    print("Get your token from: https://dashboard.ngrok.com/get-started/your-authtoken")
else:
    ngrok.set_auth_token(NGROK_AUTH_TOKEN)
    
    port = 8000
    public_url = ngrok.connect(port).public_url
    
    print("\n" + "=" * 60)
    print("✓ ngrok is ready!")
    print("=" * 60)
    print(f"API URL: {public_url}/classify")
    print(f"Web UI: {public_url}/")
    print("=" * 60)
    print("\nProceed to the next cell to launch the server.")
```

```python
# 4. Launch the FastAPI server
# This cell will run continuously while the server is active.
# Stop this cell when you're done testing.

import nest_asyncio
nest_asyncio.apply()
import uvicorn
import app

print("=" * 60)
print("Launching FastAPI Server")
print("=" * 60)

config = uvicorn.Config(app.app, host='0.0.0.0', port=8000, log_level='info')
server = uvicorn.Server(config)
await server.serve()
```
